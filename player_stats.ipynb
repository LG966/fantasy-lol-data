```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/items-stats.csv')
df.drop(columns=['item', 'item_points'], inplace=True)
df.drop_duplicates(inplace=True)

df.head()
```

```python
df.groupby(['team', 'role']).mean().unstack().plot(kind='bar', figsize=(20, 10), title="Averge score per role per team")
```

```python
df.groupby(['opponent', 'role']).mean().unstack().transpose().plot(kind='bar', figsize=(20, 10), title='Average score per role against a specific team')
```
